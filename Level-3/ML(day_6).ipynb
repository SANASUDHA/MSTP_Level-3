# Naive Bayes

```python
import pandas as pd
#import scikit-learn dataset library
from sklearn import datasets
#load dataset
wine=datasets.load_wine()
wine
#this data set is the resultof a chemical analysis
#of wines set is the same region in Italy
#but analysis determined the quantities of 13 constituents found in each of 
```

```python
print(wine.data)
```

```python
print(wine.target)
#print(wine.target_names)
```

```python
#import train_test _split function
from sklearn.model_selection import train_test_split
#split dataset into training set and test set
X_train,X_test,y_train,y_test=train_test_split(wine.data,wine.target,test_size=0.3,random_state=15)
```

```python
#import gaussian naive baye model
from sklearn.naive_bayes import GaussianNB

#Create a gaussian classifier
gnb=GaussianNB()

#Train the model using the training sets
gnb.fit(X_train,y_train)
#predict the response for test dataset
y_pred=gnb.predict(X_test)
y_pred
```

```python
#import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model A, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
print("CM",metrics.confusion_matrix(y_test,y_pred))
#accuracy_score(y_test,y_pred)
#c=confusion_matrix(y_test,y_pred)
```

 # K Nearest Neighbour 

```python
import numpy as np
import pandas as pd
data=pd.read_csv('T-shirt.csv')
data
```

```python
x=(161-data['Height(in cms)'])**2+(61-data['Weight(in kgs)'])**2
x
```

```python
data['distance']=np.sqrt(x)
```

```python
data
```

```python
data.columns
```

```python
X=data[['Height(in cms)','Weight(in kgs)']]
X
```

```python
y=data[['T-shirt size']]
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y=LabelEncoder()
Y=labelencoder_y.fit_transform(y)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
#import knearest neighbors classifier model 
from sklearn.neighbors import KNeighborsClassifier
#create KNN Classifier
knn=KNeighborsClassifier(n_neighbors=5)
# Train the model using the training sets
knn.fit(X_train,y_train)
#predict the response for test dataset
y_pred=knn.predict(X_test)

#import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
#Model Accuracy, how often is the classifier correct?
print("Accuracy",metrics.accuracy_score(y_test,y_pred))
```

```python
import numpy as np
import pandas as pd
bird=pd.read_csv('bird.csv')
#bd
bird.head()
```

```python
#bd.columns
#bd.replace(to_replace=['huml', 'humw', 'ulnal', 'ulnaw', 'feml', 'femw', 'tibl', 'tibw',
      # 'tarl', 'tarw'],value=0)
bird=bird.replace(to_replace=np.nan,value=0)
```

```python
bird.isna().sum()
```

```python
X=bird.iloc[:,1:11].values
Y=bird.iloc[:,11:].values
#X,Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=10)
X_train,Y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
rfc_predict=rfc.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(Y_test,rfc_predict)
```
