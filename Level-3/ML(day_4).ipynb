### Classifications

- Studies how to automatically learn to make accurate predictions based on past obseravtions.
- In the terminology of machine learning, learning where a training set of correctly identified observations is available.
- Machine learning cycle includes the following steps:
   - **Planning**
   - **Data Preparation**
   - **Modelling**
   - **Model Presentation**

- **Classification** is the problem of identifying to which set of categories(sub-populations) a new observation belongs.
- On the basis of a training set of data containing observations (or instances) whose category membership is known.


### common classification algorithms:

- Logistic regression(Linear Classifier-output is weighted sum of inputs)
- Naive bayes Classifier
- Support vector machine
- Decision tree
- Random Forest
- K-Nearest Neighbour

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
titanic_data = pd.read_csv("titanic.csv")
titanic_data.head(10)
```

```python
print("No of passangers in original data:" +str(len(titanic_data.index)))
```

```python
sns.countplot(x="Survived",hue="Sex",data=titanic_data)
```

```python
sns.countplot(x="Survived",hue="Pclass",data=titanic_data)
```

```python
titanic_data["Age"].plot.hist()
```

```python
titanic_data.isnull().sum()
```

```python
sns.boxplot(x="Pclass",y="Age",data=titanic_data)
```

```python
titanic_data.drop("Cabin",axis=1,inplace=True)
titanic_data.head(5)
```

```python
titanic_data.isnull().sum()
```

```python
sex=pd.get_dummies(titanic_data['Sex'])
sex.head(5)
```

```python
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark.head(5)
```

```python
Pc1=pd.get_dummies(titanic_data["Pclass"],drop_first=True)
Pc1.head()
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,Pc1],axis=1)
#titanic_data
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket','Pclass'],axis=1,inplace=True)
```

```python
titanic_data.drop('Age',axis=1,inplace=True)
```

```python
X=titanic_data.drop("Survived",axis=1) # independent variable
y=titanic_data["Survived"]
```

```python
from sklearn.model_selection import train_test_split #give model_selection
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)  # calculate accuracy
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score (y_test,predictions)*100)
```
